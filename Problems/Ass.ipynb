```python
import numpy as np 
import matplotlib.pyplot as plt 
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
def SHM(t0, tf, h, x0, v0, r, m, k, b, F0, w0):
    
    t = np.arange(t0, tf+h, h)  # Initial Values
    x = x0                      # Position
    v = v0                      # Velocity
    w = 2*np.pi*r/v0            # Angular Frequency
    
    x_values = []
    v_values = []
    
    def f(t, x, v):
        return v
    
    def g(t, x, v):
        return -w**2*x - (b/m)*v + (F0/m)*np.sin(w0*t)
    
    for i in range(len(t)):
        
        x_values.append(x)
        v_values.append(v)
        
        k1 = h*f(t[i], x, v)
        l1 = h*g(t[i], x, v)
        k2 = h*f(t[i]+ h/2, x+ k1/2, v+ l1/2)
        l2  = h*g(t[i]+ h/2, x+ k1/2, v+ l1/2)
        k3 = h*f(t[i]+ h/2, x+ k2/2, v+ l2/2)
        l3 = h*g(t[i]+ h/2, x+ k2/2, v+ l2/2)
        k4 = h*f(t[i]+ h, x+ k3, v+ l3)
        l4 = h*g(t[i]+ h, x+ k3, v+ l3)
        
        x = x + (k1 + 2*k2 + 2*k3 + k4)/6
        v = v + (l1 + 2*l2 + 2*l3 +l4)/6
        
    X, V = np.array(x_values), np.array(v_values)
    
    fig, axes = plt.subplots(2,2, figsize=(12,12))

    ax = axes[0][0]
    ax.plot(t, X, label="Position x(t)")
    ax.plot(t,V , label="Velocity v(t)")
    ax.set_xlabel("Time(t)")
    ax.set_ylabel("Position(x) & Velocity(v)")
    ax.legend(loc="upper right", fontsize=12)

    fig.text(0.5, 0.9, f"Oscillations ( w = {w:.2f}, m = {m}, b = {b:.2f}, $w_{0}$ = {w0:.2f}, $F_{0}$ = {F0} )", ha='center', fontsize=20)

    ax = axes[0][1]
    K = (1/2)*m*V**2
    P = (1/2)*k*X**2
    ax.plot(X, K, label="K.E(x)")
    ax.plot(X, P , label="P.E(x)")
    ax.set_xlabel("Position(x)")
    ax.set_ylabel("Energy(x)")
    ax.legend(loc="upper center", fontsize=12)
    
    ax = axes[1][0]
    ax.plot(t, K, label="K.E(t)")
    ax.plot(t, P , label="P.E(t)")
    ax.set_xlabel("Time(t)")
    ax.set_ylabel("Energy(t)")
    ax.legend(loc="upper right", fontsize=12)
    
    # plt.savefig("Oscillation.png", dpi=200)
    plt.show()
```

```python
SHM(0, 10, 0.01, 0, 2, 1, 1, 10, 0, 0, 0)
```
